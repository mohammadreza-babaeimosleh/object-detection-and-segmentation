## Import

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Define Functions

```python
def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

## Q1

### Contour Matching

```python
Image1 = cv2.imread('./AerialView.jpeg')
imshow("Aerial View Image",Image1)
```

```python
Image_Gray = cv2.cvtColor(Image1, cv2.COLOR_BGR2GRAY)
imshow("Gray Scale", Image_Gray)
```

```python
BilateralFilter = cv2.bilateralFilter(Image_Gray, 9, 75, 75)
low_th = 0
high_th = 80
BilateralCanny = cv2.Canny(BilateralFilter, low_th, high_th)
imshow("Bilateral Canny Smooth Filter", BilateralCanny)
```

```python
kernel = np.ones((5,5),np.uint8)
Dilation = cv2.dilate(BilateralCanny, kernel, iterations = 6)
imshow("Bilateral Canny Smooth Filter", Dilation)
```

```python
contours, hierarchy = cv2.findContours(image = cv2.bitwise_not(Dilation), mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_NONE)
Image_Contour = cv2.drawContours(image=Image1.copy(), contours=contours, contourIdx=-1, color=(255, 0, 0), thickness=10, lineType=cv2.LINE_AA)
imshow("Contour", Image_Contour)
```

### Sorting By Area

```python
copy = Image1.copy()
sorted_contours = sorted(contours, key=cv2.contourArea, reverse=True)
for (i,c) in enumerate(sorted_contours):
    M = cv2.moments(c)
    if M['m00'] > 12000:
        cx = int(M['m10'] / M['m00'])
        cy = int(M['m01'] / M['m00'])
        cv2.putText(copy, str(i), (cx, cy), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 0), 5)
        cv2.drawContours(copy, [c], -1, (255,0,0), 10)
imshow('Contours by area', copy)
```
